### Column-wise and multiple function application

```python
tips = pd.read_csv('tips.csv')
tips
```

```python
tips.groupby('sex').mean()
```

```python
tips.groupby('sex').std()
```

```python
tips.groupby('sex').agg(['mean','std', 'count']) #ara agrupar
```

```python
tips.groupby(['sex', 'smoker']).agg([np.mean, np.std, np.count_nonzero]) #introducirlo como lista
```

```python
def minmax(series):
    return series.max() - series.min()
```

```python
tips.groupby(['sex', 'smoker']).agg(minmax)
```

```python
tips.groupby(['sex', 'smoker']).agg({'total_bill':'mean', 'size':np.count_nonzero})
```

## Group-wise operations and transformations

```python
smoker_gb = tips.groupby('smoker')['total_bill'].agg([np.mean, 'std'])
smoker_gb
```

```python
tips.merge(smoker_gb, right_index=True, left_on="smoker") #el que esta en la izquierda el merge lo hace por smoker 
```

### Apply: General split-apply-combine

```python
tips['tip_pct'] = tips['tip']/tips['total_bill'] #ha dado el nombe tip_pct a la columna tip
tips
```

```python
def top(df, n=5, col='tip_pct'):
    return df.sort_values(by=col, ascending=False).head(n)

# coge la columna porcentaje de propias ordena de mayor propina da en porcentaje
```

```python
tips_by_smoker = tips.groupby('smoker')['tip_pct'].agg([np.mean, np.std])
tips_by_smoker
```

#### Suppressing the group keys

```python
tips.groupby(['sex', 'smoker']).apply(top)
```

```python
tips.groupby(['sex', 'smoker'], group_keys=False).apply(top) 
```

### Quantile and bucket analysis

```python
merged = tips.groupby(['sex', 'smoker'], group_keys=False).apply(top) 
```

```python
pd.cut(merged['total_bill'], 5) #merged invetnado
```

```python
tips.groupby(pd.cut(merged['total_bill'], 5)).count()
```
