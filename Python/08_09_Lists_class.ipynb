## Sort by custom function

__Anonimous or lambda functions__  
simple functions consisting of a single statement, the result of which is the return value.  
defined using the lambda keyword, which has no meaning other than “we are declaring an anonymous function.”  



```python
def short_function(x):
    return x *2

the_same_short_function=lambda x: x*2
```

**lambda** {VARIABLE}: {RETURN (VARIABLE}}

```python
type(the_same_short_function)
```

```python
short_function(4)
```

```python
the_same_short_function(2)
```

```python
def apply_to_list(some_list, some_function):
    results=[]
    for value in some_list:
        results.append(some_function(value))
    return results

ints=[1, 2, 3, 4]
apply_to_list(ints, lambda x:x*2)
```

adding **for comprehension** we further reduce code

```python
def apply_to_list2(some_list, some_function):
    return([some_function(x) for x in some_list]) 
ints=[1, 2, 3, 4]
apply_to_list2(ints, lambda x:x*3)
```

# Functions as objects

__function is used as argument to other function  
ops has a list of the operations to apply to a particular set of values__


```python
def add_one(value):
    return value+1
def double_value(value):
    return 2*value
def add_three(value):
    return value+3

math_ops =[add_one, double_value, add_three]
```

math_ops is a list of functions!

Now we can define a function that takes a list of functions as input :)

```python
def math_values(values,ops):
    result = []
    for value in values:
        for function in ops:
            my_value = function(value)
            result.append(my_value)
    return result
            
```

```python
k=[1,2,3]
k_res=math_values(k,math_ops)
k_res
```

We can put this in the same list using append

```python
k2=[]
for value in range(3):
    k2.append(k_res[value*3:(value+1)*3])
```

```python
k2
```

Of course, we can give lambda functions...

```python
math_values(k,[lambda x:x+1,lambda z:z*2, lambda y:y+3])
```


These are techniques to apply function to element, column or dataframe.

* **Map**: It iterates over each element of a series.
df[‘column1’].map(lambda x: 10+x), this will add 10 to each element of column1.
df[‘column2’].map(lambda x: ‘AV’+x), this will concatenate “AV“ at the beginning of each element of column2 (column format is string).

* **Apply**: As the name suggests, applies a function along any axis of the DataFrame.
df[[‘column1’,’column2’]].apply(sum), it will returns the sum of all the values of column1 and column2.

* **ApplyMap**: This helps to apply a function to each element of dataframe.
func = lambda x: x+2
df.applymap(func), it will add 2 to each element of dataframe (all columns of dataframe must be numeric type)

# List comprehensions

* allow to concisely form a new list by filtering the elements of a collection and transforming the elements passing the filter in one concise expression.


```python
a=[1,2,3,4, 5, 6]
result = []
for val in a:
    if val%2:
        result.append(val**2)
    else:
        result.append(1)
        
result
```

```python
a
```

```python
if (4%2): print("hola")
```

```python
result2=[val**2 for val in a if (val%2)]
result2
```

```python
result3=[val**2 if val%2 else 1 for val in a ] #--> changes the place for if condition
result3
```

```python
def transform(x):
    if x==1: return 0
    elif x%2: return x**2
    else: return 1

result4=[transform(val) for val in a ]
result4
```

```python
result5=map(transform,a)
list(result5)
```

```python
strings=['a', 'as', 'bat', 'car', 'dove','python']
[x.upper() for x in strings if len(x)>2]
```

# Nested list Comprehension

* nested list comprehensions are a bit hard to wrap your head around. 
* The for parts of the list comprehension are arranged according to the order of nesting, 
* filter condition is put at the end as before. 
* example
  * we “flatten” a list of tuples of integers into a simple list of integers:



```python
some_tuples=[(1,2,3),(4,5,6),(7,8,9)]
flatten=[]
for tup in some_tuples:
    for element in tup:
        flatten.append(element)
flatten
```

```python
some_tuples=[(1,2,3),(4,5,6),(7,8,9)]
#[ return_element for_loops_for_coming_to_element ]
flatten2=[element for tup in some_tuples for element in tup]
flatten2
```

  * Keep in mind that the order of the for expressions would be the same if you wrote a nested for loop instead of a list comprehension
