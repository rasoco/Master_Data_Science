# ZIP FUNCTION

“pairs” up the elements of a number of lists,tuples, or other sequences, to create a sequence of tuples

__pairing is ended when the shortest sequence is exhausted__

returns a zip object which is in fact an iteretor

```python
seq1=[1, 2, 3, 4, 5, 6]
seq2=(7, 8, 9, 10)
seq3=[True, False, True]
```

```python
type(zip(seq1, seq2, seq3))
```

```python
table=tuple(zip(seq1, seq2, seq3))
table
```

```python
table=list(zip(seq1, seq2, seq3))
table
```
