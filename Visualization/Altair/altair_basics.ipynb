```
hist_brain = alt.Chart(brain).mark_bar().encode(
    x=alt.X('Brain Weight', bin=alt.Bin(maxbins=100)), #alt.X dar mas informacion
    y='count()'
).properties(
    title='Distribución del peso de los cerebros',
    width=400
).interactive()
```

```
hist_brain
```

```
hist_body = alt.Chart(brain).mark_bar().encode(
    x=alt.X('Body Weight', bin=alt.Bin(maxbins=100)),
    y='count()'
).properties(
    title='Distribución del peso de los cuerpos',
    width=400
).interactive()
```

```
brain.columns
```

```
scatter_brain_body = alt.Chart(brain).mark_point().encode(
    x='Brain Weight',
    y='Body Weight'
).properties(
    title='Distribución del peso de los cerebros y cuerpos',
    width=800
).interactive()
```

```
scatter_brain_body
```

```
(hist_body|hist_brain)&scatter_brain_body
```

```
hist_brain = alt.Chart(brain).mark_bar().encode(
    x=alt.X('Brain Weight', bin=alt.Bin(maxbins=100)), #alt.X dar mas informacion
    y='count()'
).properties(
    title='Distribución del peso de los cerebros',
    width=400
).interactive()

hist_brain

hist_body = alt.Chart(brain).mark_bar().encode(
    x=alt.X('Body Weight', bin=alt.Bin(maxbins=100)),
    y='count()'
).properties(
    title='Distribución del peso de los cuerpos',
    width=400
).interactive()

hist_body

scatter_brain_body = alt.Chart(brain).mark_point().encode(
    x='Brain Weight',
    y='Body Weight'
).properties(
    title='Distribución del peso de los cerebros y cuerpos',
    width=800
).interactive()

scatter_brain_body

brain_dash = (hist_body|hist_brain)&scatter_brain_body
brain_dash.save('brain_dash.html') #To save HTML

```

```
select_points = alt.selection(type="interval",encodings=["x","y"])

hist_brain = alt.Chart(brain).mark_bar().encode(
    x=alt.X('Brain Weight',bin=alt.Bin(maxbins=10)),
    y="count()"
).properties(
    title="Distribución del peso de los cerebros",
    width=400
).transform_filter(
    select_points
)
hist_brain

hist_body = alt.Chart(brain).mark_bar().encode(
    x=alt.X('Body Weight',bin=alt.Bin(maxbins=10)),
    y="count()"
).properties(
    title="Distribución del peso de los cuerpos",
    width=400
).transform_filter(
    select_points
)
hist_body

scatter_brain_body = alt.Chart(brain).mark_point().encode(
    x='Brain Weight',
    y='Body Weight'
).properties(
    title="Distribución del peso de los cerebros",
    width=800,
    selection=select_points
)
scatter_brain_body

scatter_brain_body&(hist_body|hist_brain)
```

### Trends Dataset

```
trends = pd.read_csv("https://raw.githubusercontent.com/rezpe/datos_viz/master/google_trends.csv")
```

```
trends.head()
```

```
trends.columns
```

```
comp_trends = alt.Chart(trends).mark_bar().encode(
    x='search_term',
    y='mean(value)',
    color='search_term'
)
```

```
line_trends = alt.Chart(trends).mark_line().encode(
    x='yearmonth(date):T', #especifica que es una fecha. Para coger solo el año "year(date):T"
    y='mean(value)',
    color='search_term'
  )
```

```

comp_trends = alt.Chart(trends).mark_bar().encode(
    x='search_term',
    y='mean(value)',
    color='search_term'
).properties(
    width=100
)
comp_trends

line_trends = alt.Chart(trends).mark_line().encode(
    x='yearmonth(date):T', #especifica que es una fecha. Para coger solo el año "year(date):T"
    y='mean(value)',
    color='search_term'
  ).properties(
      width=600
  )

line_trends

unificate = (comp_trends|line_trends).properties(title='Busquedas de Google en términos de Data Science')
unificate
unificate.save("unificate.html")
```

```
select_st = alt.selection(type="interval",encodings=["x"])

comp_trends = alt.Chart(trends).mark_bar().encode(
    x='search_term',
    y='mean(value)',
    color=alt.condition(select_st, 'search_term', alt.value('lightgray')),
    tooltip="search_term"
).properties(
    width=100,
    selection=select_st
)

trends_line = alt.Chart(trends).mark_line().encode(
    x="yearmonth(date):T",
    y='mean(value)',
    color="search_term"
).properties(
    width=600
).transform_filter(
    select_st
)

(comp_trends|trends_line).properties(
    title="Búsquedas de Google de términos de data science"
)

```

```
select_time = alt.selection(type="interval",encodings=["x"])

comp_trends = alt.Chart(trends).mark_bar().encode(
    x='search_term',
    y='mean(value)',
    color='search_term'
).properties(
    width=100
).transform_filter(
    select_time
)

trends_line = alt.Chart(trends).mark_line().encode(
    x="yearmonth(date):T",
    y='mean(value)',
    color="search_term"
).properties(
    width=600,
    selection=select_time
)

(comp_trends|trends_line).properties(
    title="Búsquedas de Google de términos de Data Science"
)

```

```
select_time = alt.selection(type="interval",encodings=["x"])

trends_line = alt.Chart(trends).mark_line().encode(
    x="yearmonth(date):T",
    y='mean(value)',
    color="search_term"
).properties(
    width=600,
    height=400
).transform_filter(
    select_time
)

trends_line_zoom = alt.Chart(trends).mark_line().encode(
    x="yearmonth(date):T",
    y='mean(value)',
    color="search_term"
).properties(
    width=600,
    height=100,
    selection=select_time
)

trends_line&trends_line_zoom
```

### Weather Dataset

```
weather_df = pd.read_csv("https://github.com/rezpe/datos_viz/blob/master/worldTemperatures.csv?raw=true", header=None)
```

```
weather_df.head()
```

```
weather_df.columns.astype(str)
```

```
weather_df.columns = ['0', '1', '2', 'date', '4', 'temp', '6', '7', '8', '9', '10', 'lat', 'lon',
       'country', 'city']
```

```
alt.Chart('weather_df').mark_bar().encode(
    x='country',
    y='mean(temp)'
)
#error porque es muy grande el fichero
```

```
#Agrupar para que funcione
temp = weather_df.groupby("country").mean().reset_index()
alt.Chart(temp).mark_bar().encode(
    x=alt.X("country", 
            sort=alt.Sort(field='temp', 
                          op='mean', 
                          order='descending'),
                          axis=None
            ),
    y="mean(temp)",
    tooltip='country'
).properties(
    width=400
)
```

```
select_ctry = alt.selection(type="single",encodings=["x"])

temp = weather_df.groupby(["city","country"]).mean().reset_index()

ranking_city = alt.Chart(temp).mark_bar().encode(
    x=alt.X("city",
            sort=alt.Sort(field="temp",
                          op="mean",
                          order="descending"),
            
            ),
    y="mean(temp)",
    tooltip="city"
).properties(
    width=800,
    title="Ranking de ciudades por temperatura media"
).transform_filter(
    select_ctry
)

ranking_country = alt.Chart(temp).mark_bar().encode(
    x=alt.X("country",
            sort=alt.Sort(field="temp",
                          op="mean",
                          order="descending"),
            axis=None
            ),
    y="mean(temp)",
    tooltip="country"
).properties(
    width=800,
    title="Ranking de ciudades por temperatura media",
    selection=select_ctry
)

ranking_country&ranking_city
```

```
weather_df['date']=pd.to_datetime(weather_df['date'], format='%Y%m%d')
```

```
temp = weather_df.groupby('date').mean().reset_index()
alt.Chart(temp).mark_line().encode(
    x='date',
    y='mean(temp)'
)
```

```
temp = weather_df.groupby(["date","country"]).mean().reset_index()
#temp = temp[~temp["date"].dt.strftime("%Y-%m-%d").isin(["2017-04-10"])]

select_ctry = alt.selection(type="single",encodings=["x"])

ranking_country = alt.Chart(temp).mark_bar().encode(
    x=alt.X("country",
            sort=alt.Sort(field="temp",
                          op="mean",
                          order="descending"),
            axis=None
            ),
    y="mean(temp)",
    tooltip="country"
).properties(
    width=800,
    title="Ranking de ciudades por temperatura media",
    selection=select_ctry
)

evo_weather = alt.Chart(temp).mark_circle().encode(
    x="date:T",
    y="mean(temp)"
).properties(
    width=800
).transform_filter(
    select_ctry
)

ranking_country&evo_weather
```
